```python
import math


class Player:
    def __init__(self, id_n, name, runs, balls, fours, sixes, strike):
        self.id_n = id_n
        self.name = name
        self.runs = runs
        self.balls = balls
        self.fours = fours
        self.sixes = sixes
        self.strike = strike


class team(Player):
    def __init__(self, score, wicket, team_players, on_ground):
        self.team_score = score
        self.wicket = wicket
        self.team_players = team_players
        self.on_ground = on_ground

    def acceptBattingOrder(self, tot_players):
        for p in range(tot_players):
            name = input()
            # First two players are on Ground and First Player is on Strike
            if p == 0 or p == 1:
                self.on_ground.append(Player(p, name, 0, 0, 0, 0, p))

            self.team_players.append(Player(p, name, 0, 0, 0, 0, 0))

    def printScoreboard(self):
        print("Name Runs  4s   6s Balls")

        for batsman in self.team_players:
            # Not out players
            if batsman.id_n == self.on_ground[0].id_n or batsman.id_n == self.on_ground[1].id_n:
                print(batsman.name, "*  ", end=" ")
            else:
                print(batsman.name, "   ", end=" ")

            print(batsman.runs, " ", batsman.fours, "   ",
                  batsman.sixes, "   ", batsman.balls, "   ", sep=' ')

    def acceptOverRuns(self, target):
        balls_left = 6
        on_strike_id = self.changeStrike()
        while balls_left > 0:
            ball = input()

            # If Runs
            if ball.isnumeric():
                ball = int(ball)
                self.team_score += ball
                self.team_players[on_strike_id].runs += ball
                self.team_players[on_strike_id].balls += 1

                if ball == 3 or ball == 1:
                    on_strike_id = self.changeStrike()
                elif ball == 4:
                    self.team_players[on_strike_id].fours += 1
                elif ball == 6:
                    self.team_players[on_strike_id].sixes += 1

                balls_left -= 1

                # taret achieved. Stop the inning
                if self.team_score >= target:
                    break

            # If Wide
            elif ball == "Wd":
                self.team_score += 1

            # If No Wicket
            elif ball == "W":
                self.wicket += 1

                # ALL OUT. Stop the inning
                if self.wicket == len(self.team_players) - 1:
                    break
                if self.wicket < len(self.team_players):
                    next_player_id = max(self.on_ground[0].id_n, self.on_ground[1].id_n) + 1
                    if self.on_ground[0].id_n == on_strike_id:
                        self.on_ground[0] = self.team_players[next_player_id]
                        self.on_ground[0].strike = 1
                    else:
                        self.on_ground[1] = self.team_players[next_player_id]
                        self.on_ground[1].strike = 1

                    on_strike_id = next_player_id
                balls_left -= 1

    def changeStrike(self):
        strike_id = self.on_ground[1].id_n if self.on_ground[0].strike == 1 else self.on_ground[0].id_n
        self.on_ground[0].strike, self.on_ground[1].strike = self.on_ground[1].strike, self.on_ground[0].strike
        # print("Changing Strike to ", strike_id)
        return strike_id


if __name__ == '__main__':

    print("The Game Begins\n_________________")
    Overs = int(input("Number of Overs:"))
    tot_players = int(input("Number of Players:"))

    print("Batting Order for team 1")
    team_one = team(0, 0, [], [])
    team_one.acceptBattingOrder(tot_players)
    print("___________________")
    for i in range(Overs):
        print("\nOver", i + 1)
        team_one.acceptOverRuns(math.inf)  # No target for first inning

        # If ALl Out, Stop the inning
        if team_one.wicket == tot_players - 1:
            team_one.printScoreboard()
            print("TOTAL:", team_one.team_score, "/", team_one.wicket)
            break;

        print("Team one after ", i + 1, "Overs")
        team_one.printScoreboard()
        print("TOTAL:", team_one.team_score, "/", team_one.wicket)
    
    #SECOND INNING
    print("\n_____________________\n_____________________\n")
    print("Batting Order for team 2")
    team_two = team(0, 0, [], [])
    team_two.acceptBattingOrder(tot_players)
    print("___________________")
    for i in range(Overs):
        print("\nOver", i + 1)
        team_two.acceptOverRuns(team_one.team_score + 1)

        if team_two.team_score >= team_one.team_score + 1:
            team_two.printScoreboard()
            print("TOTAL:", team_two.team_score, "/", team_two.wicket)
            break

        if team_two.wicket == tot_players - 1:
            team_two.printScoreboard()
            print("TOTAL:", team_two.team_score, "/", team_two.wicket)
            break

        print("Team two after ", i + 1, "Overs")
        team_two.printScoreboard()
        print("TOTAL:", team_two.team_score, "/", team_two.wicket)

    print("\n_____________________\n_____________________\n")
    print("RESULT")
    if team_one.team_score > team_two.team_score:
        print("Team One Won by ", team_one.team_score - team_two.team_score, " Runs")
    elif team_one.team_score < team_two.team_score:
        print("Team Two Won by ", tot_players - team_two.wicket, "Wickets")
    else:
        print("Match Draw")
```
